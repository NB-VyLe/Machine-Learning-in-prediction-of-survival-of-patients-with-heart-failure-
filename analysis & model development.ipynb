3. DATA MODELING

```python
# load packages
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.utils import shuffle
```

```python
# numbers of columns
columns = heart.shape[1]
columns
```

```python
#  features
X = heart.iloc[:, 0:columns-1]
X.shape
```

```python
# number of features
features_num = X.shape[1]
features_num
```

```python
# target
y = heart.iloc[:, columns-1]
y.shape
```

3.1 FEATURE SELECTION for KNN

```python
# chose random state for shuffling
for i in range(0, 8):
    selected_features = []
    max_score = 0.0
    random_index = shuffle(range(0, features_num), random_state = i)
    print('random state:', i)
    for curr_feature in range(0, features_num):
        
        selected_features.append(random_index[curr_feature])
        new_data = X.iloc[:, selected_features].copy()
        X_train, X_test, y_train, y_test = train_test_split(new_data, y, random_state = 2)
        clf = KNeighborsClassifier(n_neighbors = 6, weights = 'distance', p =1)
        fit = clf.fit(X_train, y_train)
        cur_score = clf.score(X_test, y_test)
        if cur_score < max_score :
            selected_features.remove(random_index[curr_feature])
        else:
            max_score = cur_score
            
            print('score with', len(selected_features), 'selected features:', cur_score)
    print()
```

```python
# chose random state for Training, with shuffle random state = 4
for i in range(0, 8):
    selected_features = []
    max_score = 0.0
    random_index = shuffle(range(0, features_num), random_state = 4)
    print('random state:', i)
    for curr_feature in range(0, features_num):
        
        selected_features.append(random_index[curr_feature])
        new_data = X.iloc[:, selected_features].copy()
        X_train, X_test, y_train, y_test = train_test_split(new_data, y, random_state = i)
        clf = KNeighborsClassifier(n_neighbors = 6, weights = 'distance', p =1)
        fit = clf.fit(X_train, y_train)
        cur_score = clf.score(X_test, y_test)
        if cur_score < max_score :
            selected_features.remove(random_index[curr_feature])
        else:
            max_score = cur_score
            
            print('score with', len(selected_features), 'selected features:', cur_score)
    print()
```

```python
# shuffle random state = 4, traing random state =4
selected_features = []
max_score = 0.0
random_index = shuffle(range(0, features_num), random_state = 4)

for curr_feature in range(0, features_num):
    selected_features.append(random_index[curr_feature])
    new_data = X.iloc[:, selected_features].copy()
    X_train, X_test, y_train, y_test = train_test_split(new_data, y, random_state = 2 )
    clf = KNeighborsClassifier(n_neighbors = 6, weights = 'distance', p =1)
    fit = clf.fit(X_train, y_train)
    cur_score = clf.score(X_test, y_test)
    if cur_score < max_score :
        selected_features.remove(random_index[curr_feature])
    else:
        max_score = cur_score
        print('score with', len(selected_features), 'selected features:', cur_score)
```

3.2 TRAIN MODEL

3.2.1 KNeighborsClassifier

```python
# choose random state
for i in range(0, 10):
    X_train, X_test, y_train, y_test = train_test_split(df, y,random_state = i)
    clf = KNeighborsClassifier(6)
    fit = clf.fit(X_train, y_train)
    y_predicted = fit.predict(X_test)
    print('random state =' ,i)
   
    print(confusion_matrix(y_test, y_predicted))
    print(clf.score(X_test, y_test))
    print()
    
```

```python
# choose K with random state at 2
for i in range(3, 16):
    X_train, X_test, y_train, y_test = train_test_split(df, y, random_state = 2)
    clf = KNeighborsClassifier(i)
    fit = clf.fit(X_train, y_train)
    y_predicted = fit.predict(X_test)
    print('K =' ,i)
    print(confusion_matrix(y_test, y_predicted))
    print(clf.score(X_test, y_test))
    print()
```

```python
# Traing random state = 2, KNN k =9, and default for others parameters

X_train, X_test, y_train, y_test = train_test_split(df, y,random_state = 2)
clf = KNeighborsClassifier(9)
fit = clf.fit(X_train, y_train)
y_predicted= fit.predict(X_test)
print(confusion_matrix(y_test, y_predicted))
print(clf.score(X_test, y_test))
print(classification_report(y_test, y_predicted))
```

```python
#using Training random state = 2, n_neighbors = 6, weights = 'distance', p =1
# the best result for KNN
X_train, X_test, y_train, y_test = train_test_split(df, y,random_state = 2)
clf = KNeighborsClassifier(n_neighbors = 6, weights = 'distance', p =1)
fit = clf.fit(X_train, y_train)
y_predicted= fit.predict(X_test)
print(confusion_matrix(y_test, y_predicted))
print(clf.score(X_test, y_test))
print(classification_report(y_test, y_predicted))
print('Table 3: Classification report for KNeighbors classifier')
```

3.2.2 FEATURE SELECTION - FOR DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# find the random state to shuffle
for i in range(0, 10):
    selected_features = []
    max_score = 0.0
    random_index = shuffle(range(0, features_num), random_state = i)
    print('random state:', i)
    for curr_feature in range(0, features_num):
        
        selected_features.append(random_index[curr_feature])
        new_data = X.iloc[:, selected_features].copy()
        X_train, X_test, y_train, y_test = train_test_split(new_data, y, random_state = 0)
        clf = DecisionTreeClassifier(  criterion='entropy',min_samples_split =2, min_samples_leaf = 7)
        fit = clf.fit(X_train, y_train)
        cur_score = clf.score(X_test, y_test)
        if cur_score < max_score :
            selected_features.remove(random_index[curr_feature])
        else:
            max_score = cur_score
            
            print('score with', len(selected_features), 'selected features:', cur_score)
    print()
```
